```python
import pandas as pd
```

```python
# Descarga del archivo
url = "https://s3.opensky-network.org/data-samples/metadata/aircraft-database-complete-2025-01.csv"
df = pd.read_csv(url, on_bad_lines="skip")
```

```python
df.columns = [col.replace("'", "") for col in df.columns]
```

```python
columnas = ["icao24", "engines", "manufacturerName", "model", "operator"]
df = df[columnas]
```

```python
def clean_quotes(df):
    return df.map(lambda x: np.nan if x == "''" else str(x).replace("'", ""))
```

```python
import numpy as np
```

```python
df = clean_quotes(df)
```

```python
def clean_engines(engine_str):
    
    if pd.isna(engine_str):
        return np.nan
    
    engine_str = engine_str.strip()
    num_engines = int(engine_str[0]) if engine_str and engine_str[0].isdigit() else np.nan
    
    return num_engines

# Aplicar la función de limpieza
df['num_engines'] = df['engines'].apply(lambda x: clean_engines(x))
```
